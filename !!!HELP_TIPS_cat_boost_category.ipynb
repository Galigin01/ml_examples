# Человеческое обучение для категориальных значений

```python
# разобьем % выходов по количеству продуктов у юзеров (исходя из Phik)
train_full.groupby('NumOfProducts')['Exited'].agg(['count','mean'])

# получается что этот процент есть вероятность оттока у группы юзеров
# это будет наш скоринг оттока по группам
```

```python
# создадим из этого модель
# добавим индекс, а старый индекс превратим в группы с кол-вом продуктов
model = train_full.groupby('NumOfProducts')['Exited'].mean().reset_index()
model = model.rename({'Exited':'score_prod'}, axis=1)
model
```

```python
# мерджим модель с таблицей, что бы добавить скоринг оттока
train_full = train_full.merge(model, how='left', on = 'NumOfProducts')
```

```python
# Baseline
# добавим соеднее по таблице
train_full['score_mean'] = train_full['Exited'].mean()
```

```python
train_full.head(5)
```

```python
train_full.score_prod.mean()
```

```python
# log_loss - функция логорифмической потери
# это общий показатель точности в логистической регрессии.
# Чем ниже значение log loss, тем точнее прогнозирование
# В идеальной модели значение log loss было бы равно 0.
from sklearn.metrics import log_loss
```

```python
# посмотрим где log_loss меньше
display(log_loss(train_full['Exited'],train_full['score_mean'])) # baseline
display(log_loss(train_full['Exited'],train_full['score_prod'])) # скоринг/вероятность
```

```python
# Uplift - метрика показывающая во сколько раз лучше средний показатель целевой переменной,
# чем baseline по нужному топу

# foo - это средние топ 20%, отсротированных по score_prod и их среднее значение целевой переменной
foo = train_full.sort_values(by='score_prod', ascending=False).head(int(len(train_full) * 0.2))['Exited'].mean()

# voo - это среднее по общей целевой переменной
voo = train_full['Exited'].mean()
display(foo)
display(voo)
f"uplift = {foo/voo}"

# Разделив средний показатель Exited всего датасета на Exited топ-20 мы получим uplift
# метрика интуитивная, но при более точном предикте при неоднородном датасете разрыв должен увеличиваться
```

```python
# Сделаем функцию для uplift
def uplift(df, score, prc):
 foo = df.sort_values(by=score, ascending=False).head(int(len(df) * prc))['Exited'].mean()
 voo = df['Exited'].mean()
 return foo/voo
```

```python
uplift(train_full, 'score_prod', 0.2)
```

```python
# смерджим тест с моделью и рсчитаем те же метрики для тест test
test = test.merge(model, how='left', on='NumOfProducts')
```

```python
display(uplift(test, 'score_prod', 0.2))
display(log_loss(test['Exited'],test['score_prod']))
```

# Age
Добавим в модель условие по возрасту


```python
# разобьем по равным частям train
train_full['Age_group'] = pd.qcut(train_full['Age'],5)
```

```python
# посмотрим зависимости
train_full.groupby('Age_group')['Exited'].agg(['count','mean'])

# зависимость есть: средний Exited (отток) возрастает с возрастом
```

```python
# теперь используем разбиение qcat для cat (так как в тесте уже не будет возможности делить на квантили)
train_full['Age_group'] = pd.cut(train_full['Age'],[0,18, 31, 35, 40, 46, float('inf')]) # прибьем значения cut-ом
train_full
```

```python
# создадим модель с группировкой по 2 параметрам
model = train_full.groupby(['NumOfProducts','Age_group'])['Exited'].mean().reset_index()
model = model.rename({'Exited':'score_prod_age'}, axis=1) # переименуем
model.head(10)
```

```python
# сджойним модель и данные
train_full = train_full.merge(model, how='left', on = ['NumOfProducts', 'Age_group'])
```

```python
train_full.head(10)
```

```python
# uplift увеличился т.е увеличился разрыв между средним значением топ(20) и средним всего датасета
# (значит данные менее однородны => более индивидуальны и различны предикты)
print('uplift')
display(uplift(train_full, 'score_prod_age', 0.2))
display(uplift(train_full, 'score_prod', 0.2))
print('log_loss')
# посмотрим где log_loss меньше, log_loss меньше значит лучше (идеал 0)
display(log_loss(train_full['Exited'],train_full['score_mean'])) # baseline
display(log_loss(train_full['Exited'],train_full['score_prod'])) # скоринг/вероятность score_prod_
display(log_loss(train_full['Exited'],train_full['score_prod_age'])) # скоринг/вероятность score_prod_age
```
